### Question 2

```python
lower = []
for i in corpus:
    lower.append(' '.join([word.lower() for word in i.split()]))
alpha = []
for i in lower:
    alpha.append(' '.join([word for word in i.split() if word.isalpha()]))
alpha[0]
```

```python
tokenize = []
from nltk.tokenize import word_tokenize
for i in alpha:
    tokenize.append(word_tokenize(i))
```

```python
import nltk
from nltk.corpus import *
stopword = nltk.corpus.stopwords.words('english')
no_stop = []
for i in tokenize:
    no_stop.append([word for word in i if word not in stopword])
```

```python
final = []
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for i in no_stop:
    final.append(' '.join([ps.stem(word) for word in i]))
```

#### From Scratch

> a. Binary

```python
doc_count={}
for i in range(len(final)):
    word_count={}
    for word in final[i].split():
        if word not in word_count.keys():
            word_count[word]=1
    doc_count[i]=word_count
```

```python
binary_tdm_scratch=pd.DataFrame(doc_count)
binary_tdm_scratch.fillna(0,inplace=True)
binary_tdm_scratch
```

> b. Actual Term Frequency

```python
doc_count1 = {}
for i in range(len(final)):
    word_count1 = {}
    for word in final[i].split():
        if word not in word_count1.keys():
            word_count1[word] = 0
        word_count1[word]+=1
    doc_count1[i] = word_count1
```

```python
tdm_tf_scratch=pd.DataFrame(doc_count1)
tdm_tf_scratch.fillna(0,inplace=True)
tdm_tf_scratch
```

> c. Term Frequency with length normalization

```python
tdm_length_norm_scratch=tdm_tf_scratch.iloc[:,:].div(tdm_tf_scratch.sum(axis=0),axis=1)
tdm_length_norm_scratch
```

> d. Term Frequency-Inverse Document Frequency

```python
no_of_docs=100
unique_terms=tdm_length_norm_scratch.index
idf_count={} 
for term in unique_terms:
    idf_count[term]=0 
    for doc in final:
        if term in doc.split():
            idf_count[term]+=1
    idf_count[term]=np.log10(no_of_docs/idf_count[term])
```

```python
idf_series=pd.Series(idf_count)
idf=pd.DataFrame(idf_series)
idf
```

```python
tf_idf_scratch=tdm_length_norm_scratch*idf.values
tf_idf_scratch
```

### Question 4

> a. Term Co-occurence Matrix

```python
co_occur_matrix=np.dot(binary_tdm_scratch.values,binary_tdm_scratch.values.T)
np.fill_diagonal(co_occur_matrix,0)
co_occur_matrix=pd.DataFrame(co_occur_matrix, index=binary_tdm_scratch.index, columns=binary_tdm_scratch.index)
co_occur_matrix
```

> b. Positive Pointwise Mutual Information (PPMI)

```python
terms_occur_mat = np.zeros(co_occur_matrix.shape[0])
for i in range(len(terms_occur_mat)):
    terms_occur_mat[i] = sum(binary_tdm_scratch.iloc[i])
```

```python
ppmi_matrix = co_occur_matrix
for i in range(co_occur_matrix.shape[0]):
    n_i = terms_occur_mat[i]
    for j in range(co_occur_matrix.shape[1]):
        if(i==j):
            continue
        val = co_occur_matrix.iloc[i][j]
        n_j = terms_occur_mat[j]
        ppmi_matrix.iloc[i][j] = max(np.log10(val/(n_i * n_j)), 0)
```
